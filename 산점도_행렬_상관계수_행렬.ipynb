# #산점도 행렬 

- 두 개 이상의 변수가 있는 데이터에서 변수들 간의 산점도를 그림 

1. 대각선의 히스토그램을 통해 이상치를 확인함
2. 종속변수와 설명변수들의 관계를 시각적으로 판단 
3. 종속변수가 수치형이면 각 설명변수와의 직선 상관관계를 비교
4. 종속변수가 범주형인 경우 종속변수를 잘 구분하는 변수를 파악 
5. 설명변수간의 직선 함수를 통해 다중공산성 문제를 파악 



````python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.datasets import load_iris

iris=load_iris()
iris=pd.DataFrame(iris.data,columns=iris.feature_names)
iris['Class']=load_iris().target
iris['Class']=iris['Class'].map({0:'Setosa',1:'Versicolour',2:'Virginica'})
````

````python
scatter_matrix(iris,alpha=0.5,figsize=(8,8),diagonal='hist')
plt.show()
````

````python
#타겟별 차이가 나는 데이터에 대해서는 target범주별 색깔을 다르게 지정하여 산점도 행렬 
import seaborn as sns
sns.pairplot(iris,diag_kind='auto',hue='Class')
plt.show()
````

#상관계수 행렬 그래프 
1. corr()함수를 통해서 상관행렬 생성 
2. data=data.corr(method='pearson)
3. 상관행렬 생성 후 seaborn의 heatmap()을 사용하여 상관행렬을 시각화한다. 


```
sns.heatmap(data,xticklabels=data.columns,yticklabels=data.columns,cmap='RdBu_r',annot=True)
```
- cmap : 히트맵의 색깔 
- annot : True일 경우 상관계수를 텍스트로 표시 


````python
iris_corr=iris.drop(columns='Class').corr(method='pearson')
sns.heatmap(iris_corr,xticklabels=iris_corr.columns,yticklabels=iris_corr.columns,cmap='RdBu_r',annot=True)
plt.show()
````
